<h1> Import Data </h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
%matplotlib inline

#sat scores
sat = pd.read_csv('sat_scores.csv')
sat.columns = ['State_Abbrev','PR','Verbal','Math','Region']

#state abbreviations
abbrevs = pd.read_html('https://en.wikipedia.org/wiki/List_of_U.S._state_abbreviations', header=1)
abbrevs = abbrevs[0].set_index('Name').loc['Alabama':'Wyoming',['ANSI']]
abbrevs.index.name = 'State'
abbrevs.columns = ['State_Abbrev']

#state populations
pops = pd.read_html('https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_population')
pops = pops[1].set_index(2).iloc[1:55,[2]]
pops.index.name = 'State'
pops.columns = ['Population']

def tryint(x):
    try:
        return int(x)
    except UnicodeEncodeError:
        return float('Nan')
pops['Population'] = [tryint(x) for x in pops['Population']]

#state incomes
incomes = pd.read_html('https://en.wikipedia.org/wiki/List_of_U.S._states_by_GDP_per_capita', header = 0)
incomes = incomes[0].set_index('State').loc[:,['2015']]
incomes = incomes[incomes.loc[:,'2015']>0]
incomes.columns = ['Income']
```

<h1> Merge the Data </h1>

```python
#merge sat with abbrevs
alldata = pd.merge(sat, abbrevs.reset_index(), on='State_Abbrev').set_index('State')
alldata = pd.merge(alldata, pops, left_index=True, right_index=True)
alldata = pd.merge(alldata, incomes, left_index=True, right_index=True)
alldata['Weight'] = alldata['PR'] * alldata['Population'].apply(int) / (100 * alldata['Population'].sum())
alldata['Tot_Score'] = alldata['Verbal'] + alldata['Math']
alldata['Diff_Score'] = alldata['Verbal'] - alldata['Math']
alldata
```

<h1> SAT scores by state and region, verbal and math </h1>

```python
# Define a lambda function to compute the weighted mean:
wm = lambda x: np.average(x, weights=alldata.loc[x.index, "Weight"])

# Define a dictionary with the functions to apply for a given column:
f = {'Verbal': {'weighted_mean' : wm}, 'Math': {'weighted_mean' : wm} }

# Groupby and aggregate with your dictionary:
q1 = alldata.groupby(["Region"]).agg(f)

q1.plot(kind='bar')
```

```python
q2 = alldata.loc[:,['Tot_Score','Region']]
q2 = q2[['Tot_Score','Region']].sort_values(by='Tot_Score')
fig, (ax1,ax2,ax3,ax4) = plt.subplots(1,4,figsize=(20,10),sharey=True)
q2[q2.Region=='South'].plot(kind='bar', ax=ax1)
q2[q2.Region=='West'].plot(kind='bar', ax=ax2)
q2[q2.Region=='Midwest'].plot(kind='bar', ax=ax3)
q2[q2.Region=='Northeast'].plot(kind='bar', ax=ax4)
```

<h1> Participation rate by state and region </h1>

```python
q2 = alldata.loc[:,['PR','Region']].sort_values(by='PR')
fig, (ax1,ax2,ax3,ax4) = plt.subplots(1,4,figsize=(20,10),sharey=True)
q2[q2.Region=='South'].plot(kind='bar', ax=ax1)
q2[q2.Region=='West'].plot(kind='bar', ax=ax2)
q2[q2.Region=='Midwest'].plot(kind='bar', ax=ax3)
q2[q2.Region=='Northeast'].plot(kind='bar', ax=ax4)
```

<h1> Relative Student Population by state and region </h1>

```python
q2 = alldata.loc[:,['Weight','Region']].sort_values(by='Weight')
fig, (ax1,ax2,ax3,ax4) = plt.subplots(1,4,figsize=(20,10),sharey=True)
q2[q2.Region=='South'].plot(kind='bar', ax=ax1)
q2[q2.Region=='West'].plot(kind='bar', ax=ax2)
q2[q2.Region=='Midwest'].plot(kind='bar', ax=ax3)
q2[q2.Region=='Northeast'].plot(kind='bar', ax=ax4)
```
