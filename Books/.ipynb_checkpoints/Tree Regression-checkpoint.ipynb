```python
import pandas as pd
import numpy as np
import sklearn.cross_validation as cv
import sklearn.linear_model as lm
import matplotlib.pyplot as plt
%matplotlib inline

# Read in realEstate.csv as a data frame
reale = pd.read_csv('3.2-lab/assets/datasets/Sacramentorealestatetransactions.csv')
```

```python
split = cv.train_test_split
(train_X,test_X,train_Y,test_Y) = split(reale, reale.price,test_size=0.1)

ols = lm.LinearRegression()
ols.fit(train_X[['beds','baths','sq__ft']],train_Y)

zip(['beds','baths','sq__ft'],ols.coef_)
```

```python
predictions = ols.predict(test_X[['beds','baths','sq__ft']])
plt.scatter(test_Y,predictions)
```

```python
import sklearn.metrics
sklearn.metrics.r2_score(test_Y,predictions)
```

```python
#preprocess the type into something that can be a regressor
from sklearn import preprocessing
label = preprocessing.LabelEncoder()
reale['type'] = label.fit_transform(reale['type'])

dropcols = ['sale_date','city','state','street','price']

import sklearn.tree
tree = sklearn.tree.DecisionTreeRegressor()
tree.fit(train_X.drop(dropcols,1),train_Y)
tree_predictions = tree.predict(test_X.drop(dropcols,1))
plt.scatter(test_Y,tree_predictions)

sklearn.metrics.r2_score(test_Y,tree_predictions)
```
