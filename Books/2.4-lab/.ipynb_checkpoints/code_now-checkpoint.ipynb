```python
import pandas as pd, numpy as np
bill = pd.read_csv("../../Books/2.4-lab/assets/datasets/billboard.csv",parse_dates=['date.entered','date.peaked'])
bill = bill.sort_values(by='artist.inverted')

def clean(x):
    try:
        if np.isnan(x):
            return 0
    except TypeError:
        pass
    return x

bill = bill.applymap(clean)
#bill['date.entered'] = pd.to_datetime(bill['date.entered'])
#bill['date.peaked'] = pd.to_datetime(bill['date.peaked'])

colnames = list(bill.columns)
for i in range(1,77):
    if i < 10:
        colnames[6+i] = 'Week' + '0' + str(i)
    else:
        colnames[6+i] = 'Week' +str(i)
bill.columns = colnames

bill
#bill.ix[0,'date.peaked']
```

```python
def mean(x):
    return x.mean()
def std(x):
    return x.std()
def q0(x): 
    return x.quantile(0)
def median(x):
    return x.quantile(0.5) 
def q100(x):
    return x.quantile(1)

aggs = {'sum':sum,
        'mean':mean#,
        #'std':std,
        #'min':q0,
        #'median':median, 
        #'max':q100
       } 
cols = [x for x in bill if x.startswith('Week')]
bill.groupby(['genre'])[cols].agg(aggs).T.unstack(0)
```

```python
pv = pd.pivot_table(bill, index=['genre'], values=cols, aggfunc=[median, mean, std]).T
pv[(pv.T !=0).any()].unstack(0)
```
