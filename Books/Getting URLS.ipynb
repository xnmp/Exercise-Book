<h1>find all \<a\> tags</h1>

```python
from bs4 import BeautifulSoup
wiki = requests.get('https://en.wikipedia.org/wiki/States_and_territories_of_Australia')
soup = BeautifulSoup(wiki.text,'lxml')
#this finds the number of links
len(soup.find_all('a'))
```
