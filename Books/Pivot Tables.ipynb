```python
def threshold_count(xs,thresh=100):
    tot = 0
    for x in xs:
        if x>thresh:
            tot += 1
    return tot
    

trains = pd.read_csv('sydtrains.csv',thousands=',')
pd.pivot_table(trains, 
               index=['LINE','YEAR'], 
               values=['IN_24_HOURS','OUT_24_HOURS'],
               #columns=['YEAR'],
               aggfunc=[sum,len,threshold_count])
```
